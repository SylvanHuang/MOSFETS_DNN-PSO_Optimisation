## Imports

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

## Data reading and preparation

```python
# reading given excel file
filename = 'Delays_W_L_FAMIRRORSimulated_50K.xlsx'
df = pd.read_excel(filename)
df.head(100)
```

```python
# divide into test and train set
data = df.values
np.random.shuffle(data)
train = data[:40000, :]
test = data[40000:, :]
print(train.shape)
```

```python
# converting into polynomial features
train_x = train[:, :56]
train_y = train[:, 56:]

test_x = test[:, :56]
test_y = test[:, 56:]

poly = PolynomialFeatures(degree=1)
train_x_poly = poly.fit_transform(train_x)
test_x_poly = poly.fit_transform(test_x)
```

## Training and testing

```python
for i in range(6):
  reg = LinearRegression().fit(train_x, train_y[:, i])
  print('Train score for output variable %d = %f' % (i, reg.score(train_x, train_y[:, i])))
  print('Test  score for output variable %d = %f' % (i, reg.score(test_x, test_y[:, i])))
  print()
```
